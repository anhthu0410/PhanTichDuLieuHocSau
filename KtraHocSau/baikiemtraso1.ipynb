```python
import pandas as pd
import numpy as np
```

```python
#Câu 1: Tạo dataframe
names = np.array(['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ivan', 'Jack', 'Kelly', 'Liam', 'Mona', 'Nina', 'Oscar'])
ages = np.array([25, 30, 35, 28, 22, 45, 34, 31, 27, 29, 33, 40, 26, 32, 36])
salaries = np.array([50000, 60000, 70000, 55000,52000,80000,72000,68000,61000,59000,63000,77000,53000,66000,75000])
data = pd.DataFrame({'name': names,'age': ages,'salary': salaries})


```

```python
#Câu 2: Hiển thị thông tin về DataFrame
print(data)
```

```python
#Câu 3: Lọc các hàng có giá trị lớn hơn 28 trong cột salary
data[data['age'] >= 28]
data.groupby('name')['name'].agg('count')
```

```python
#Câu 4: Tính giá trị tb của cột salary
mean_salary = np.mean(data['salary'].values)
print(mean_salary)
```

```python
#Câu 5: Nhóm dữ liệu theo cột age và tính tổng salary cho mỗi nhóm
data.groupby('age')['salary'].sum().reset_index()
print(data)
```

```python
#Câu 6: sắp xếp dataframe theo cột salary giảm dần
data.sort_values(by="salary", ascending=False)
print(data)
```

```python
#Câu 7: Vẽ biểu đồ cho cột age
cau7 =data.groupby('age')['age'].agg('count')
cau7.plot.bar()
```

```python
#Câu 8: vẽ biểu đồ đường cho cột salary
data['salary'].plot()
```

```python
#Câu 9: vẽ biểu đồ tròn cho cột age
data['age'].plot.pie()
```

```python
#Câu 10: Vẽ biểu đồ phân tán cho age và salary
import matplotlib.pyplot as plt
correlation_dh1_t1 = data[['age', 'salary']].corr()
print(correlation_dh1_t1)
plt.scatter(data['age'], data['salary'])
```

```python
#Câu 11: Kiểm tra xem có giá trị NaN nào trong dataframe không
data.isnull().sum()
```

```python
#Câu 12: thay thế các giá trị ở cột age lớn hơn 30 bằng giá trị tb của cột đó
tuoitb = data['age'].mean()

data.loc[data['age'] > 30, 'age'] = tuoitb

print(data)
```

```python
#Câu 13: Chuẩn hóa cột age về khảong giá trị từ 0 đến 1
data['age_chuanhoa'] = (data['age'] - data['age'].min()) / (data['age'].max() - data['age'].min())

print(data)
```

```python
#Câu 14: tạo cột mới age_group phân loại thành young, middle-age và old
def categorize_age(age):
    if age < 30:
        return 'young'
    elif 30 <= age <= 60:
        return 'middle-age'
    else:
        return 'old'

data['age_group'] = data['age'].apply(categorize_age)
print(data)
```

```python
#Câu 15: tính tỷ lệ % thay đổi của cột salary
data['salary_tltd'] = data['salary'].pct_change() * 100

print(data)
```

```python
#câu 16: tìm các giá trị trùng lập trong cột name và loại bỏ các hàng trùng lập, giữ lại hàng đầu tiên
data.drop_duplicates(subset='name', keep='first')

print(data)
```

```python
#Câu 17: lưu data thành file csv
data.to_csv('bài kiểm tra số 1.csv', index=False)
```
