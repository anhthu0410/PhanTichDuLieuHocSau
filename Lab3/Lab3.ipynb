```python
column_names = ["Id", "Name", "Age", "Weight", 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218']
df = pd.read_csv("patient_heart_rate.csv", names = column_names)
print(df.head())
```

```python
#Problem 2
df[['Firstname','Lastname']] = df['Name'].str.split(expand=True)
df = df.drop('Name', axis=1)
print (df)
```

```python
#Problem 3
weight = df['Weight']

for i in range (0, len(weight)):
    x = str(weight[i])

    if "lbs" in x[-3:]:

        x = x[:-3:]

        float_x = float(x)
        
        y = int(float_x/2.2)

        y = str(y)+"kgs"
        weight[i] = y
print(df)

```

```python
#Problem 4
df.dropna(how="all", inplace=True)
print(df)
```

```python
#Problem 5
df = df.drop_duplicates(subset=['Firstname','Lastname', 'Age', 'Weight'])
print(df)
```

```python
#Problem 6
df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex = True, inplace =True)
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex = True, inplace =True)
print(df)
```

```python
#Problem 7
df.columns = ["Id", "Age", "Weight", 'm0006', 'm0612', 'm1218', 'f0006', 'f0612', 'f1218','Firstname', 'Lastname']

# Extract relevant columns
data_relevant = df[['Age', 'Weight']]

# Convert 'Weight' to numeric, handling different units and formats
data_relevant.loc[:, 'Weight'] = pd.to_numeric(data_relevant['Weight'].str.replace('[^\d.]', '', regex=False), errors='coerce')

# Convert to numpy array
data_array = data_relevant.to_numpy()

# Identify rows where both Age and Weight are missing
rows_to_keep = ~np.isnan(data_array).all(axis=1)

# Filter out rows where both Age and Weight are missing
data_array = data_array[rows_to_keep]

# Calculate mean of Age ignoring NaNs
mean_age = np.nanmean(data_array[:, 0])

# Replace NaN values in Age with the mean
data_array[np.isnan(data_array[:, 0]), 0] = mean_age

# Convert back to DataFrame
data_cleaned_np = pd.DataFrame(data_array, columns=['Age', 'Weight'])
print(df)
```

```python
# Problem 8
df = pd.melt(df, id_vars=['Id', 'Age', 'Weight', 'Firstname', 'Lastname'], value_name="PulseRate", var_name="sex_and_time").sort_values(['Id', 'Age', 'Weight', 'Firstname', 'Lastname'])

# Extract and split values from "sex_and_time" column
tmp_df = df["sex_and_time"].str.extract(r"([mf])(\d{4})-(\d{4})", expand=True)

# Name columns
tmp_df.columns = ["Sex", "Time_lower", "Time_upper"]

# Convert "m" and "f" to "male" and "female"
tmp_df["Sex"] = tmp_df["Sex"].map({"m": "male", "f": "female"})

# Merge with original DataFrame
df = pd.concat([df, tmp_df], axis=1)

# Drop unnecessary columns
df = df.drop(['sex_and_time', 'Time_lower', 'Time_upper'], axis=1)

# Print and save to CSV
print(df.head())  # Print to verify
df.to_csv('outputcleanup.csv', index=False)
```

```python
#Problem 9
# Function to handle missing blood pressure data
def fill_missing_bp(data):
    for i in range(len(data)):
        if pd.isna(data.at[i, 'PulseRate']):
            if i > 0 and not pd.isna(data.at[i-1, 'PulseRate']):
                data.at[i, 'PulseRate'] = data.at[i-1, 'PulseRate']
            elif i < len(data)-1 and not pd.isna(data.at[i+1, 'PulseRate']):
                data.at[i, 'PulseRate'] = data.at[i+1, 'PulseRate']
            elif i > 1 and not pd.isna(data.at[i-2, 'PulseRate']):
                data.at[i, 'PulseRate'] = data.at[i-2, 'PulseRate']
            elif i < len(data)-2 and not pd.isna(data.at[i+2, 'PulseRate']):
                data.at[i, 'PulseRate'] = data.at[i+2, 'PulseRate']
            elif data['PulseRate'].notna().any():
                data.at[i, 'PulseRate'] = data['PulseRate'].mean()
            else:
                data.at[i, 'PulseRate'] = 120  # Example stable medical value
    return data

# Fill missing BloodPressure values
df = fill_missing_bp(df)
print(df)
```

```python
#Problem 10
# Select relevant columns
df = df[['Id', 'Age', 'Weight', 'Firstname', 'PulseRate']]  # Adjust based on actual columns

# Drop rows with any remaining missing values
df = df.dropna()

# Reindex the DataFrame
df.reset_index(drop=True, inplace=True)

# Save the cleaned data
cleaned_file_path = 'patient_heart_rate_clean.csv'
df.to_csv(cleaned_file_path, index=False)

cleaned_file_path
```
